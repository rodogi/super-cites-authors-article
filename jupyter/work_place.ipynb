# Add the workplace to the people table

To add the workplace to the people table, we use the workplace-id column.

```python
import pandas as pd
```

```python
people = pd.read_csv('/home/rdora/femec/data/people.csv')
work = pd.read_csv('/home/rdora/femec/data/institution.csv')
```

```python
work_short = work.rename(columns={'Handle': 'Workplace-Institution',
                                  'alpha-3': 'country_code'})
work_short = work_short[['Workplace-Institution',
                         'city',
                         'country_code',
                         'region',
                         'sub-region']]
```

```python
people = pd.merge(people,
                 work_short,
                 on='Workplace-Institution',
                 how='left')
```
